```python
class Burger:
    def __init__(self):
        self.bun = "No Bun"
        self.meat = "No Meat"
        self.salad = "No Salad"
        self.cheese = "No Cheese"
        self.sauce = "No Sauce"

    def set_bun(self, bun):
        self.bun = bun

    def set_meat(self, meat):
        self.meat = meat

    def set_salad(self, salad):
        self.salad = salad

    def set_cheese(self, cheese):
        self.cheese = cheese

    def set_sauce(self, sauce):
        self.sauce = sauce

    def print(self):
        print(f"Burger is finished!\nBun: {self.bun} - Meat: {self.meat} - Cheese: {self.cheese} - Salad: {self.salad} - Sauce: {self.sauce}")

class BurgerBuilder:
    def __init__(self):
        self.burger = Burger()

    def build_bun(self):
        pass

    def build_meat(self):
        pass

    def build_salad(self):
        pass

    def build_cheese(self):
        pass

    def build_sauce(self):
        pass

    def build(self):
        self.build_bun()
        self.build_meat()
        self.build_salad()
        self.build_cheese()
        self.build_sauce()
        return self.burger

class CheeseBurgerBuilder(BurgerBuilder):
    def build_bun(self):
        self.burger.set_bun("White Bread")

    def build_meat(self):
        self.burger.set_meat("Beef")

    def build_salad(self):
        self.burger.set_salad("Iceberg")

    def build_cheese(self):
        self.burger.set_cheese("American Cheese")

    def build_sauce(self):
        self.burger.set_sauce("Secret Sauce")

class BurgerRestaurant:
    def __init__(self):
        self.burger_builder = None

    def set_builder(self, burger_builder):
        self.burger_builder = burger_builder

    def build_burger(self):
        burger = self.burger_builder.build()
        return burger

def main():
    burger_restaurant = BurgerRestaurant()
    burger_restaurant.set_builder(CheeseBurgerBuilder())
    burger = burger_restaurant.build_burger()
    burger.print()

if __name__ == "__main__":
    main()

```
