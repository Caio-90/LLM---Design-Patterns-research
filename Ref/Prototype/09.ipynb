```python
# ==============================================================================================================================
# OBS: A página apresentava exemplos em python e java, porém o código python não funciona, este é uma conversão do exemplo java
# ==============================================================================================================================
from abc import ABC, abstractmethod

class Vehicle(ABC):
    # Class Variables
    def __init__(self, seats=0, tyres=0, color="", fuel=""):
        self.seats = seats
        self.tyres = tyres
        self.color = color
        self.fuel = fuel

    def __copy__(self):
        return self.__class__(self.seats, self.tyres, self.color, self.fuel)

    def __str__(self):
        return f"Seats: {self.seats}\nTyres: {self.tyres}\nColor: {self.color}\nFuel: {self.fuel}\n"

    @abstractmethod
    def clone(self):
        pass

class Bike(Vehicle):
    def __init__(self, seats=0, tyres=0, color="", fuel=""):
        super().__init__(seats, tyres, color, fuel)

    def clone(self):
        return Bike(self.seats, self.tyres, self.color, self.fuel)

class Car(Vehicle):
    def __init__(self, seats=0, tyres=0, color="", fuel=""):
        super().__init__(seats, tyres, color, fuel)

    def clone(self):
        return Car(self.seats, self.tyres, self.color, self.fuel)

def main():
    vehicles = []
    vehicles_copy = []

    b1 = Bike(1, 2, "Black", "Petrol")
    vehicles.append(b1)

    c1 = Car(5, 4, "White", "Electric")
    vehicles.append(c1)

    for vehicle in vehicles:
        vehicles_copy.append(vehicle.clone())

    print("Original Objects:")
    for vehicle in vehicles:
        print(vehicle)
        print("---------------------------------------------------------")

    print("\n\nPrototyped Objects:")
    for vehicle in vehicles_copy:
        print(vehicle)
        print("---------------------------------------------------------")

if __name__ == "__main__":
    main()
```
