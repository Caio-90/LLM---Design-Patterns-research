```python
import copy

class Product:
    def clone(self):
        return copy.deepcopy(self)

class Book(Product):
    def __init__(self):
        self.description = None
        self.sku = None
        self.num_pages = None

    def set_description(self, description):
        self.description = description

    def set_sku(self, sku):
        self.sku = sku

    def set_num_pages(self, num_pages):
        self.num_pages = num_pages

    def get_description(self):
        return self.description

    def get_sku(self):
        return self.sku

    def get_num_pages(self):
        return self.num_pages

class DVD(Product):
    def __init__(self):
        self.description = None
        self.sku = None
        self.duration = None

    def set_description(self, description):
        self.description = description

    def set_sku(self, sku):
        self.sku = sku

    def set_duration(self, duration):
        self.duration = duration

    def get_description(self):
        return self.description

    def get_sku(self):
        return self.sku

    def get_duration(self):
        return self.duration

class PrototypeExample:
    def __init__(self):
        self.product_map = {}

    def get_product(self, product_code):
        cached_product = self.product_map.get(product_code)
        return cached_product.clone()

    def load_cache(self):
        # for each product run expensive query and instantiate
        # product_map[product_key] = product

        # for exemplification, we add only two products
        b1 = Book()
        b1.set_description("Oliver Twist")
        b1.set_sku("B1")
        b1.set_num_pages(100)
        self.product_map["B1"] = b1

        d1 = DVD()
        d1.set_description("Superman")
        d1.set_sku("D1")
        d1.set_duration(180)
        self.product_map["D1"] = d1

def main():
    pe = PrototypeExample()
    pe.load_cache()

    cloned_book = pe.get_product("B1")
    print("SKU =", cloned_book.get_sku())
    print("Description =", cloned_book.get_description())
    print("Number of Pages =", cloned_book.get_num_pages())

    cloned_dvd = pe.get_product("D1")
    print("SKU =", cloned_dvd.get_sku())
    print("Description =", cloned_dvd.get_description())
    print("Duration =", cloned_dvd.get_duration())

if __name__ == "__main__":
    main()
    
```
