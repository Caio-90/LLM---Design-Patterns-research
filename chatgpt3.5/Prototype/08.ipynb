```python
import copy

class Address:
    def __init__(self, street, city, state, zip_code):
        self.street = street
        self.city = city
        self.state = state
        self.zip_code = zip_code
    
    def __str__(self):
        return f"{self.street}, {self.city}, {self.state}, {self.zip_code}"

class EmployeeRecord:
    def __init__(self, emp_id, name, designation, salary, address):
        self.emp_id = emp_id
        self.name = name
        self.designation = designation
        self.salary = salary
        self.address = address
    
    def show_record(self):
        print(f"ID: {self.emp_id}")
        print(f"Name: {self.name}")
        print(f"Designation: {self.designation}")
        print(f"Salary: {self.salary}")
        print(f"Address: {self.address}")

    def clone(self):
        return copy.deepcopy(self)

# Função para solicitar dados do usuário e criar um EmployeeRecord
def create_employee_record_from_input():
    print("Informe os detalhes do funcionário:")
    emp_id = input("ID: ")
    name = input("Nome: ")
    designation = input("Cargo: ")
    salary = float(input("Salário: "))
    street = input("Rua: ")
    city = input("Cidade: ")
    state = input("Estado: ")
    zip_code = input("CEP: ")

    address = Address(street, city, state, zip_code)
    return EmployeeRecord(emp_id, name, designation, salary, address)

# Exemplo de uso:

# Criando um EmployeeRecord a partir de entrada do usuário
emp1 = create_employee_record_from_input()

# Clonando o EmployeeRecord
emp2 = emp1.clone()

# Modificando o endereço do funcionário clonado
emp2.address.street = "Av. Principal"

# Exibindo os registros para verificar a clonagem
print("\nRegistro do Funcionário Original:")
emp1.show_record()

print("\nRegistro do Funcionário Clonado:")
emp2.show_record()
```
