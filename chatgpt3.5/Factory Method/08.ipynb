```python
# Interface para o tradutor de mensagens de transporte
class TransportMessageTranslator:
    def translate(self, message):
        pass

# Implementações do tradutor de mensagens de transporte para diferentes idiomas
class FrenchTransportMessageTranslator(TransportMessageTranslator):
    def translate(self, message):
        if message == "car":
            return "voiture"
        elif message == "motorcycle":
            return "moto"
        elif message == "bicycle":
            return "vélo"
        else:
            return "unknown"

class EnglishTransportMessageTranslator(TransportMessageTranslator):
    def translate(self, message):
        if message == "car":
            return "car"
        elif message == "motorcycle":
            return "motorcycle"
        elif message == "bicycle":
            return "bicycle"
        else:
            return "unknown"

class SpanishTransportMessageTranslator(TransportMessageTranslator):
    def translate(self, message):
        if message == "car":
            return "coche"
        elif message == "motorcycle":
            return "motocicleta"
        elif message == "bicycle":
            return "bicicleta"
        else:
            return "desconocido"

# Factory Method para criar instâncias baseadas na localização
class Localizer:
    def create_translator(self, location):
        if location == "França":
            return FrenchTransportMessageTranslator()
        elif location == "Inglaterra":
            return EnglishTransportMessageTranslator()
        elif location == "Espanha":
            return SpanishTransportMessageTranslator()
        else:
            raise ValueError(f"Unsupported location: {location}")

# Exemplo de uso do Factory Method
if __name__ == "__main__":
    localizer = Localizer()

    # Definindo a localização como França
    location = "França"
    translator = localizer.create_translator(location)

    # Traduzindo mensagens para a língua local
    messages = ["car", "motorcycle", "bicycle"]
    translated_messages = [translator.translate(message) for message in messages]

    print(f"Traduções para {location}:")
    for original, translated in zip(messages, translated_messages):
        print(f"{original}: {translated}")

    print()

    # Definindo a localização como Inglaterra
    location = "Inglaterra"
    translator = localizer.create_translator(location)

    # Traduzindo mensagens para a língua local
    translated_messages = [translator.translate(message) for message in messages]

    print(f"Traduções para {location}:")
    for original, translated in zip(messages, translated_messages):
        print(f"{original}: {translated}")

    print()

    # Definindo a localização como Espanha
    location = "Espanha"
    translator = localizer.create_translator(location)

    # Traduzindo mensagens para a língua local
    translated_messages = [translator.translate(message) for message in messages]

    print(f"Traduções para {location}:")
    for original, translated in zip(messages, translated_messages):
        print(f"{original}: {translated}")
```
